```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
data=pd.read_csv(r"D:\machine_learning\data\titanic_cleaned.csv")
data.shape
```

```python
data.head()
```

### Seperating indepentdent and dependent variable

```python
data.columns
```

```python
feature=data.drop("Survived",axis=1)
target=data["Survived"]
feature.shape,target.shape
```

### Spltting data into train and test set

```python
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y=train_test_split(feature,target,random_state=56)
```

### Normalize the data

```python
#importing min max scaler
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
```

```python
cols=train_x.columns
cols
```

```python
train_x_scaled=scaler.fit_transform(train_x)
train_x_scaled=pd.DataFrame(train_x_scaled,columns=cols)
train_x_scaled.head()
```

```python
test_x_scaled=scaler.fit_transform(test_x)
test_x_scaled=pd.DataFrame(test_x_scaled,columns=cols)
test_x_scaled.head()
```

## Implementing Logistic regression

```python
#importing logistic regression
from sklearn.linear_model import LogisticRegression as LR
from sklearn.metrics import f1_score
```

```python
#creating instance of Logistic regression
lr=LR()

#training the data
lr.fit(train_x_scaled,train_y)
#predicting the values
train_predict=lr.predict(train_x_scaled)
train_predict
```

```python
#calculating the f1 score
k=f1_score(train_predict,train_y)
print("f1 score of train data :",k)
```

```python
#predicting over the test data
test_predict=lr.predict(test_x_scaled)
#calculating f1 score
k=f1_score(test_predict,test_y)
print("f1 score of test data :",k)
```

#### Making prediction using predict_proba function

```python
#import predict_proba function
train_predict=lr.predict_proba(train_x_scaled)
train_predict
```

```python
train_preds=train_predict[:,1]
train_preds
```

```python
for i in range (0,len(train_preds)):
    if train_preds[i]>0.55:
        train_preds[i]=1
    else:
        train_preds[i]=0
```

```python
 #calculating f1 score
k=f1_score(train_preds,train_y)
print("Training f1_score :",k)
```

### Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
cf=confusion_matrix(test_y,test_predict)
print(cf)
```

### Classification report

```python
from sklearn.metrics import classification_report as rep
print(rep(test_y,test_predict))
```
